```python
# Kenfack Sorelle
# 2025-04-24
# P5HW Battle Game Assignment
# A simple battle game using loops, functions, and modular programming

# --- Character Management ---

def create_character():
    """
    Prompts user for character name, health, and strength.
    Returns a dictionary representing the character.
    """
    name = input("Enter character name: ")
    health = int(input("Enter character health (e.g., 100): "))
    strength = int(input("Enter character strength (e.g., 20): "))
    return {"name": name, "health": health, "strength": strength}


def display_character(character):
    """
    Displays the attributes of a character.
    Non-value returning function.
    """
    print(f"\nCharacter: {character['name']}")
    print(f"Health: {character['health']}")
    print(f"Strength: {character['strength']}\n")


def attack(attacker, defender):
    """
    Simulates an attack from one character to another.
    Reduces defender's health based on attacker's strength.
    Value-returning function. Returns the new health of the defender.
    """
    print(f"\n{attacker['name']} attacks {defender['name']} for {attacker['strength']} damage!")
    defender['health'] -= attacker['strength']
    if defender['health'] < 0:
        defender['health'] = 0
    print(f"{defender['name']}'s health is now {defender['health']}\n")
    return defender['health']


# --- Main Game Loop ---
def main():
    characters = []
    while True:
        print("\n--- Battle Game Menu ---")
        print("1. Create Character")
        print("2. Display Characters")
        print("3. Battle")
        print("4. Exit")

        choice = input("Choose an option: ")

        if choice == '1':
            char = create_character()
            characters.append(char)
            print(f"Character {char['name']} created!\n")

        elif choice == '2':
            if not characters:
                print("No characters available.")
            for char in characters:
                display_character(char)

        elif choice == '3':
            if len(characters) < 2:
                print("You need at least 2 characters to battle.")
                continue

            print("Available characters:")
            for i, char in enumerate(characters):
                print(f"{i + 1}. {char['name']}")

            try:
                idx1 = int(input("Select attacker (number): ")) - 1
                idx2 = int(input("Select defender (number): ")) - 1
                if idx1 == idx2:
                    print("A character cannot attack themselves.")
                    continue
                attack(characters[idx1], characters[idx2])
            except (IndexError, ValueError):
                print("Invalid selection.")

        elif choice == '4':
            print("Exiting game. Goodbye!")
            break

        else:
            print("Invalid option. Try again.")


if __name__ == '__main__':
    main()
```
